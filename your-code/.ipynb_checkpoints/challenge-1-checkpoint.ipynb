#### Define data.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.plot(x,
         y,            #Dividimos entre mil para ajustar eje Y
         #marker='o',         #Forma del nodo
         linestyle='dashed', #Estilo de la linea
         markersize= 8,      #Tamano del marcador  
         color='blue'       #Color del grafico
         )
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(nrows=2, #Dos filas de graficos
                       ncols=1, #Numero columnas de graficos
                       figsize=(15, 10), #Tamano del lienzo
                       sharex=True
)

ax[0].plot(x,y)                     # ax[] llama al espacio donde queremos graficar
ax[0].title.set_text('Grafico 1')
ax[0].set_xlabel('Numero')
ax[0].set_ylabel('Doble del numero')


ax[1].scatter(x,z)                     # ax[] llama al espacio donde queremos graficar
ax[1].title.set_text('Grafico 2')
ax[1].set_xlabel('Numero')
ax[1].set_ylabel('numero al cuadrado')

plt.xticks(rotation=90)


plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
fig, ax = plt.subplots(nrows=1, #Dos filas de graficos
                       ncols=2, #Numero columnas de graficos
                       figsize=(15, 10), #Tamano del lienzo
                       sharex=True
)

ax[0].plot(x,x**2,x,np.exp(x))                     
ax[0].title.set_text('Normal ')

ax[1].plot(x,x**2,x,np.exp(x))
#ax[1].set_xscale(x)
ax[1].set_yscale('log')
ax[1].title.set_text('Logaritmica')
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
df = pd.read_csv('vehicles.csv')
```

```python
df.columns
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.scatter(df['Combined MPG'],
         df['Highway MPG'],
            cmap='inferno',
            s=df['Highway MPG'],
            c=df['Combined MPG']
         )

plt.title('Combined vs. Highway MPG', fontsize=15)
plt.xlabel('Combined', fontsize=15)
plt.ylabel('Highway', fontsize=15)
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
plt.boxplot(df['CO2 Emission Grams/Mile'])
plt.title('CCO2 Emission Grams/mile', fontsize=10)
plt.ylabel('Grams/Mile', fontsize=10)
;
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
plt.hist(df['Fuel Barrels/Year'])
plt.title('Fuel Barrels/Year', fontsize=12)
plt.xlabel('Barrels/Year', fontsize=12)
plt.ylabel('Count', fontsize=12);
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
plt.bar(df['Fuel Type'],df['City MPG'])
plt.title('Fuel Type vs City MPG', fontsize=12)
plt.xlabel('Barrels/Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90);
```
